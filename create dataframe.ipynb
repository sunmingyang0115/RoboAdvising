now sort the stock list compared to 'p' based on correlation

now pick the first 10 stocks of the sorted list, with the first five to be the max weighting (which is 20% i think) and the last 5 to have miniscule weighting

now we will compute the sharpe ratio/beta of that portfolio

now we iterate through the entire portfolio, choosing every stock as the 'optimial stock' and find the portfolio that has the highest rating of sharpe ratio/beta

time complexity is O(n^2logn) which should be fast enough
